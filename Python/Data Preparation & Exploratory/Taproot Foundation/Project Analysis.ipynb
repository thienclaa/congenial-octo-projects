# Prjts

```python
prjts.head()
```

### Exploring Dataset

```python
prjts.info()
```

1) First thing I noticed is I have a lot of null values I'll have to grapple with. I can either do imputation or remove null observations/

2) I will also need to create a new column 'days_to_update' from the 'updated_at' column less the 'created_at' column. 

3) I think there're a few columns in this dataset that can be dropped.

```python
prjts = prjts.drop(labels=['needs_accomplish','needs_support','needs_value', 'partner_organization_id', 'match_job_id', 'agreed_at_community',
                          'publish_externally', 'enable_success_story', 'is_archived', 'share_metadata', 'campaign_id',
                          'image_id', 'admin_id', 'local_only', 'success_story', 'satisfaction_rating'],axis=1)
```

```python
prjts['created_at'] = pd.to_datetime(prjts['created_at'], format="%Y-%m-%d %H:%M:%S.%f")
prjts['updated_at'] = pd.to_datetime(prjts['updated_at'], format="%Y-%m-%d %H:%M:%S.%f")
prjts['days_to_update'] = (prjts['updated_at']- prjts['created_at']).dt.days
prjts.head()
```

Since the dataset is reduced, let's see how many nulls there are in each column before removing them.

```python
prjts.isnull().sum()
```

It looks like most of the null values are in project group, I will go ahead and drop those columns.

```python
prjts = prjts.drop(labels=['project_group'], axis=1)
```

```python
prjts['state'].unique()
```

```python
mapping={
    'completed': 1,
    'matched': 1,
    'applied': 1,
    'published': 1
}
prjts['succeeded'] = prjts['state'].map(mapping) == 1
prjts["succeeded"] = prjts["succeeded"].astype(int)
```

```python
prjts.sort_values(by='id',axis=0).head()
```

```python
prjts_succeed = prjts.groupby('project_category_id')['succeeded'].mean().to_frame()
prjts_succeed.sort_values(by='project_category_id').head()
```

```python
prjts_days = prjts.groupby('project_category_id')['days_to_update'].mean().to_frame()
prjts_days.sort_values(by='project_category_id').head()
```

## User_Inqry

```python
user_inqry.head()
```

### Exploring Dataset

```python
user_inqry.info()
```

```python
user_inqry.isnull().sum()
```

1) Again, I will also reduce the size here.

2) There're lot of null values but some are useful to help answering the questions such as hours or satisfaction rating as they can possibly provide some insights on volunteers of projects.

3) Although, it's not a question I'm answering but I think it would be interesting to see what the rejection rate is for volunteers. 

```python
user_inqry = user_inqry.drop(labels=['scheduled_for', 'decision_deadline', 'conference_line_id', 'hours', 'pbc_rating',
                                    'npo_rating', 'satisfaction_rating', 'pbc_review', 'archived', 'time_slots'], axis=1)
```

```python
user_inqry['created_at'] = pd.to_datetime(user_inqry['created_at'], format="mixed")
user_inqry['updated_at'] = pd.to_datetime(user_inqry['updated_at'], format="mixed")
```

```python
# Create days_to_update column
user_inqry['days_to_update'] = (user_inqry['updated_at'] - user_inqry['created_at']).dt.days
```

```python
user_inqry['state'].unique()
```

```python
mapping={
    'rejected': 'inqry_failed',
    'pbc_expired': 'inqry_failed',
    'npo_expired': 'inqry_failed',
    'cancelled': 'inqry_failed',
    'failed': 'inqry_failed',
    'admin_close': 'inqry_failed',
    'missed': 'inqry_failed'
}
user_inqry['inqry_failed'] = user_inqry['state'].map(mapping) == 'inqry_failed'
user_inqry["inqry_failed"] = user_inqry["inqry_failed"].astype(int)
```

```python
user_inqry = user_inqry.drop(labels=['created_at', 'updated_at'], axis=1)
```

```python
user_inqry.tail()
```

```python
user_inqry = user_inqry.merge(prjts[['id','project_category_id']],left_on = 'project_id',right_on='id')
user_inqry.head()
```

```python
days_to_match = user_inqry.groupby('project_category_id')['days_to_update'].mean().to_frame()
days_to_match = days_to_match.rename(columns={'days_to_update':'avg_days_to_match'})
days_to_match.sort_values(by='project_category_id').head()
```

```python
inqry_recieved = user_inqry.groupby('project_category_id')['id_x'].count().to_frame()
inqry_recieved = inqry_recieved.rename(columns={'id_x':'num_inqry_recieved'})
inqry_recieved.sort_values(by='project_category_id').head()
```

```python
match_rate = user_inqry.groupby('project_category_id')['inqry_failed'].mean().to_frame()
match_rate.sort_values(by='project_category_id').head()
```
